```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Time"], bins=50, kde=True)
plt.title("Distribution of Transaction Time")
plt.show()

fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

plt.figure(figsize=(12, 6))
sns.histplot(normal["Time"], bins=50, color="blue", label="Normal", alpha=0.4, stat="count")
sns.histplot(fraud["Time"], bins=50, color="red", label="Fraud", alpha=0.7, stat="count")
plt.legend()
plt.yscale("log")
plt.title("Fraud vs Normal Transactions Over Time (Count, log scale)")
plt.show()

```

```python
# Sắp xếp theo thời gian
df = df.sort_values(by="Time")

# Chia dữ liệu
split_index = int(len(df) * 0.8)
train_df = df.iloc[:split_index]
test_df = df.iloc[split_index:]

print("Training Set Shape:", train_df.shape)
print("Testing Set Shape:", test_df.shape)

print("\nFraud Distribution in Training Set:")
print(train_df["Class"].value_counts(normalize=True))

print("\nFraud Distribution in Testing Set:")
print(test_df["Class"].value_counts(normalize=True))
```

```python
df_sorted = df.sort_values("Time")

df_sorted["hour"] = (df_sorted["Time"] // 3600)

fraud_rate_time = df_sorted.groupby("hour")["Class"].mean()

plt.figure(figsize=(12,5))
fraud_rate_time.plot()
plt.title("Fraud Rate Over Time")
plt.xlabel("Hour")
plt.ylabel("Fraud Rate")
plt.show()
```
